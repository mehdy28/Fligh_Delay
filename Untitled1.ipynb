We will be removing the "c-" in the 'Month', 'DayofMonth', 'DayOFWeek' columns in both train and test datasets.

```python
def remove_C(data):
  return data[2:]

train['Month'] = train.apply(lambda row : remove_C(row['Month']), axis=1)
train['DayofMonth'] = train.apply(lambda row : remove_C(row['DayofMonth']), axis=1)
train['DayOfWeek'] = train.apply(lambda row : remove_C(row['DayOfWeek']), axis=1)

test['Month'] = test.apply(lambda row : remove_C(row['Month']), axis=1)
test['DayofMonth'] = test.apply(lambda row : remove_C(row['DayofMonth']), axis=1)
test['DayOfWeek'] = test.apply(lambda row : remove_C(row['DayOfWeek']), axis=1)
```

```python
#We will be replacing the 'dep_delayed_15min' with 0-1;
train['dep_delayed_15min'].replace({"Y": 1, "N": 0}, inplace=True)
```

```python
#Changing the type of the supposedly numeric columns into int64
train.DayofMonth=train.DayofMonth.astype(np.int64)
train.Month=train.DayofMonth.astype(np.int64)
train.DayOfWeek=train.DayofMonth.astype(np.int64)

test.DayofMonth=test.DayofMonth.astype(np.int64)
test.Month=test.DayofMonth.astype(np.int64)
test.DayOfWeek=test.DayofMonth.astype(np.int64)
```

## One Hot Encoder: For both train and test data

```python
#Create a list of culumns that u wish to perform encoding on
column_list = train.loc[:, (train.dtypes == "object")].columns.values.tolist()
print(column_list)
```

### One Hot Encoder: Train dataset

```python
from sklearn.preprocessing import OneHotEncoder

# Apply one-hot encoder to each column with categorical data
OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
OH_cols = pd.DataFrame(OH_encoder.fit_transform(train[column_list]),columns= (OH_encoder.get_feature_names_out() ))

# One-hot encoding removed index; put it back
OH_cols.index = train.index

# Remove categorical columns (will replace with one-hot encoding)
train = train.drop(column_list, axis=1)

# Add one-hot encoded columns to numerical features
train = pd.concat([train, OH_cols], axis=1)
```

```python
train.head()
```

### One Hot Encoder: Test dataset

```python
from sklearn.preprocessing import OneHotEncoder

# Apply one-hot encoder to each column with categorical data
OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
OH_cols = pd.DataFrame(OH_encoder.fit_transform(test[column_list]),columns= (OH_encoder.get_feature_names_out() ))

# One-hot encoding removed index; put it back
OH_cols.index = test.index

# Remove categorical columns (will replace with one-hot encoding)
test = test.drop(column_list, axis=1)

# Add one-hot encoded columns to numerical features
test = pd.concat([test, OH_cols], axis=1)
```

# Applying Machine learning models and testing

```python
#Splitting the dataset int test and train subsets
from sklearn.model_selection import train_test_split
X, y= train.drop(['dep_delayed_15min'], axis=1) ,train.dep_delayed_15min
# Divide data into training and validation subsets
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.20,random_state=0)
```

### **Logistic Regression** 

```python
from sklearn.linear_model import LogisticRegression
logisticRegr = LogisticRegression()
logisticRegr.fit(X_train, y_train)
y_pred = logisticRegr.predict(X_test)
```

### **Random Forest** 


```python
from sklearn.ensemble import RandomForestClassifier 
randFor = RandomForestClassifier(n_estimators=50, max_features="auto", random_state=0)
randFor.fit(X_train, y_train)
y_pred = randFor.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
